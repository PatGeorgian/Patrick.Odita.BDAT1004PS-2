5. Residual Analysis
Description:
Residual analysis involves evaluating the difference between observed and predicted values. It helps in checking the assumptions of the regression model, such as linearity, homoscedasticity, and normality of residuals.
Key Plots:
- Residuals vs. Fitted Values: Checks for non-linearity and homoscedasticity.
- Normal Q-Q Plot: Checks for normality of residuals.
- Scale-Location Plot: Checks for homoscedasticity.
- Residuals vs. Leverage: Identifies influential data points.


Example:
After training a linear regression model on the California Housing dataset, you can plot the residuals to check if they are randomly distributed around zero.
Exercise:
- Load the California Housing dataset.
- Train a linear regression model.
- Plot the residuals vs. fitted values.


```python
# Import necessary libraries
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the California Housing dataset
data = fetch_california_housing()
X = data.data  # Features (independent variables)
y = data.target  # Target (dependent variable)
```

```python
# Step 2: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Step 3: Train a Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Step 4: Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Step 5: Calculate residuals (difference between actual and predicted values)
residuals = y_test - y_pred
```

```python
# Step 6: Plot residuals vs. fitted values
plt.scatter(y_pred, residuals, alpha=0.5)
plt.axhline(y=0, color='red', linestyle='--')  # Add a horizontal line at y=0
plt.xlabel("Fitted Values (Predicted Prices)")
plt.ylabel("Residuals (Actual - Predicted)")
plt.title("Residuals vs. Fitted Values")
plt.show()
```
